```python
import pandas as pd
```

```python
data = pd.read_csv("yelp_boston.csv")
```

```python
boston = [];
for x in data['location']:
    if ("Boston" in x) & ("West Roxbury" not in x) & ("Dorchester" not in x) & ("Hyde Park" not in x) & ("Mattapan" not in x) & ("Roslindale" not in x)& ("Jamaica Plain" not in x):
        boston.append(x)
```

```python
data = data.loc[data['location'].isin(boston)]
```
